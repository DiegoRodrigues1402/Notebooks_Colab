```python
import pandas as pd
```

```python
url ='https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv'
dados = pd.read_csv(url,sep=';')
dados
```

soma dos valores valor +  condominio

```python
dados['valor_po_mes'] = dados['Valor'] + dados['Condominio']
```

```python
dados.head()
```

soma dos valores valor por mes + iptu

```python
dados['valor_po_ano'] = dados['valor_po_mes'] * 12 + dados['IPTU']
dados.head()
```

```python
dados['Descrição'] = (dados['Tipo'] + ' em ' + dados['Bairro'] + ' com '+
                      dados['Quartos'].astype(str) + ' quarto(s) ' +
                      ' e ' + dados['Vagas'].astype(str) + ' de garagem ')

dados.head()
```

fazer uma validação se existe suites ou não
dados

```python
dados['Posui_suites'] = dados['Suites'].apply(lambda x:'Sim'if x>0 else 'Não')
```
