```python
import pandas as pd
```

```python
url ='https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url,sep=',')
dados.head()
```

```python
dados["Pontos_extras"] = dados['Notas'] * 40 / 100
dados.head()
```

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados.head()
```

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x :"Aprovado" if x >= 6 else 'Reprovado')
dados.head(18)
```

```python
antigos_reprovados = dados.query('Aprovado == False')
antigos_reprovados
```
